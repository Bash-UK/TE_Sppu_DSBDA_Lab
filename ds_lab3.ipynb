#  Grouping 

```python

groups= df['species'].unique()
print(groups)
no = df.groupby('species').ngroups
print(f"Total groups : {no}")
```

# Group wise summary

```python
df.groupby('species').describe()
```

# Display groupwise

```python
gb = df.groupby('species')

for key,value in gb:
    print(gb.get_group(key))
```
