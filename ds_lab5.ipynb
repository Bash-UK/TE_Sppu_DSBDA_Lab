2. Show the distribution of Age with respect to gender (Histogram / Boxplot)
3. Show Survival count with respect to gender and age group (Histogram / Boxplot)

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
df1 = df.pivot_table(index='Sex' , aggfunc={'Survived':sum})
df1
```

# Male & Female Distribution according to age group [Histogram]

```python
df2 = df.groupby(['Sex'])
df_male = df2.get_group('male')
df_female = df2.get_group('female')
```

```python
sns.histplot(data=df_male, x ='Age', bins = 20, kde = True)
```

```python
sns.histplot(data=df_female, x ='Age', bins = 20, kde = True)
```

# Male Survival count according to Age groups

```python
df_male_surv_age = df_male.groupby(['Survived'])
df_male_surv_age1 = df_male_surv_age.get_group(1)
df_male_surv_age1
```

```python
sns.histplot(data=df_male_surv_age1, x ='Age', bins = 20)
```

# Female Survival count according to Age groups

```python
df_female_surv_age = df_female.groupby(['Survived'])
df_female_surv_age1 = df_female_surv_age.get_group(1)
df_female_surv_age1
```

```python
sns.histplot(data=df_female_surv_age1, x ='Age', bins = 20)
```

# Male & Female Distribution according to age group [Box plot] 

```python
sns.boxplot(data=df, x="Age", y="Sex")
```

# Gender X Survived According to Age group [Box plot]

```python
sns.boxplot(data=df, x="Age", y="Sex", hue="Survived")
```
