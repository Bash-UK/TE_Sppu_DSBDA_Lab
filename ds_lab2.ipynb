```python
q1=df7['Total'].quantile(0.25)
q2=df7['Total'].quantile(0.50)
q3=df7['Total'].quantile(0.75)
IQR=q3-q1
UL=q3+(1.5*IQR)
LL=q1-(1.5*IQR)
out= ~df7['Total'].between(LL,UL)
df7[out]
```
