```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
fig,axs = plt.subplots(2,2,figsize=(7,9))
sns.histplot(data = df, x="sepal_length", bins = 10 ,color = "skyblue", ax= axs[0,0],kde = True)
sns.histplot(data = df, x="sepal_width", bins = 10, color = "Teal", ax= axs[0,1],kde= True)
sns.histplot(data = df, x="petal_length", bins = 10, color = "olive", ax= axs[1,0],kde=True)
sns.histplot(data = df, x="petal_width", bins = 10, color = "green", ax= axs[1,1],kde=True)
```

```python
fig,axs = plt.subplots(2,2,figsize=(9,10))

sns.boxplot(data = df, x="species", y = "sepal_length"  ,color = "skyblue", ax= axs[0,0])
sns.boxplot(data = df, x="species", y = "sepal_width", color = "Teal", ax= axs[0,1])
sns.boxplot(data = df, x="species",y = "petal_length", color = "olive", ax= axs[1,0])
sns.boxplot(data = df,x= "species", y="petal_width" , color = "orange", ax= axs[1,1])
```
